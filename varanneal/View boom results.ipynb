```python
import numpy as np
import matplotlib
matplotlib.use("nbagg")
import matplotlib.pyplot as plt
import os
import pickle
```

### Find the data

```python
# results_folder = "original_data_P6"
results_folder = "noisy_original_data_P55"
results_folder2 = "original_data_P55"
data_path = "/Volumes/AJT Flash Drive/abarbanel_group/neurodyn/" + results_folder
data_path2 = "/Volumes/AJT Flash Drive/abarbanel_group/neurodyn/" + results_folder2
Ninit = 30
```

#### Plot the data

```python
fig, ax = plt.subplots(1, 2, sharey=True, figsize=(10,5))
for i in xrange(Ninit):
    try: 
        temp_data = np.load(data_path + "/action_errors%d.npy" % (i + 1))
        temp_data2 = np.load(data_path2 + "/action_errors%d.npy" % (i + 1))
        ax[0].plot(temp_data[:,1])
        ax[1].plot(temp_data2[:,1])
    except(IOError):
        pass
plt.yscale("log")
# ax[0].set_title("Action v. $R_f/R_m$, 6 estimated parameters")
ax[0].set_title("Action v. $R_f/R_m$, 55 estimated parameters, with noise")
ax[1].set_title("Action v. $R_f/R_m$, 55 estimated parameters")
plt.savefig("original_data_no_noise_action_errors")
plt.show()
```
